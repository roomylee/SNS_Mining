```python
import pandas as pd
from konlpy.tag import Twitter
```

```python
df = pd.read_csv('tweet.csv')
df
```

```python
print(len(df))
```

```python
pos_tagger = Twitter()

def tokenize(doc):
    result = list()
    for s in pos_tagger.pos(doc, norm=True, stem=True):
        if s[1] in ['Noun','Verb','Adjective']:
            result.append(s[0])
    return result
```

```python
train_docs = list()
for sentence in df['Contents']:
    try:
        train_docs.append(tokenize(sentence))
    except Exception as e:
        continue
```

```python
print(train_docs[:10])
```

```python
from gensim.models import word2vec
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
model = word2vec.Word2Vec(train_docs)
```

```python
print(model['대통령'])
print(len(model['대통령']))
print(model['대통령'].shape)
```

```python
print(model.most_similar('대통령'))
```

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
X = [model['대통령'], model['박근혜'], model['탄핵'], model['노력'], model['취업'], model['공부'], model['문재인']]
```

```python
X
```

```python
pca.fit(X)
```

```python
print(pca.explained_variance_ratio_)
```

```python
projected_X = pca.transform(X)
projected_X
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
labels = ['daetong','park','out','hard','job','study','moon']
```

```python
plt.scatter(
    projected_X[:, 0], projected_X[:, 1], marker='o',
    cmap=plt.get_cmap('Spectral'))
for label, x, y in zip(labels, projected_X[:, 0], projected_X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-20, 0),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
```
